## Barras

```python
poblacion_region_df = df.groupby(['Region'])['Poblacion 0-14','Poblacion 15-64', 'Poblacion 65+'].sum()
poblacion_region_df
```

```python
len(poblacion_region_df)
```

```python
plt.bar(range(6), poblacion_region_df['Poblacion 0-14'])
plt.xticks(range(6), list(poblacion_region_df.index), rotation='vertical')
plt.title('Poblacion 0-14')
plt.show()
```

```python
plt.bar(range(6), poblacion_region_df['Poblacion 15-64'])
plt.xticks(range(6), list(poblacion_region_df.index), rotation='vertical')
plt.title('Poblacion 15-64')
plt.show()
```

```python
plt.bar(range(6), poblacion_region_df['Poblacion 65+'])
plt.xticks(range(6), list(poblacion_region_df.index), rotation='vertical')
plt.title('Poblacion 65+')
plt.show()
```

```python
plt.figure(figsize=(15,6))
for i, c in enumerate(poblacion_region_df.columns):
    plt.subplot(1, 3, i+1)
    plt.bar(range(6), poblacion_region_df[c])
    plt.xticks(range(6), list(poblacion_region_df.index), rotation='vertical')
    plt.title(c)
```

```python
plt.figure(figsize=(15,6))
for i, c in enumerate(poblacion_region_df.columns):
    plt.subplot(1, 3, i+1)
    plt.barh(range(6), poblacion_region_df[c])
    plt.yticks(range(6), list(poblacion_region_df.index), rotation='horizontal')
    plt.title(c)
```

```python
fig, axs = plt.subplots(1, 3, sharex = 'col', sharey = 'row', figsize=(15,6))
plt.style.use('fast')
plt.subplots_adjust(wspace=0.05, hspace=0.05)
colors=['red','green','purple']
cont = 0
for i, c in enumerate(poblacion_region_df.columns):
    axs[i].barh(list(poblacion_region_df.index), poblacion_region_df[c],color=colors[cont])
    axs[i].set_xlabel(c)
    cont+=1
```

## Cajas

```python
df['PIB_CUANTILES']=pd.qcut(df['PIB'], 5)

df.boxplot(
    column='Gasto en salud ( del PIB)',
    by='PIB_CUANTILES',
    figsize=(15,10)
)
plt.show()
```

```python
df['PT_CUANTILES']=pd.qcut(df['Poblacion total'], 5)

df.boxplot(
    column='Tasa de mortalidad infantil',
    by='PT_CUANTILES',
    figsize=(15,10)
)
plt.show()
```

## pie

```python
ranges = [0, 3, 6, 9, np.inf]
names = ['0-2', '3-5', '6-8', '9+']

df['GS_PIB'] = pd.cut(
    df['Gasto en salud ( del PIB)'],
    bins=ranges,
    labels=names
)

df['GS_PIB'].value_counts().plot.pie()
plt.show()
```

```python
plt.style.use('seaborn')
fig, axs = plt.subplots(1, 3, sharex = 'col', sharey = 'row', figsize=(15,6))
plt.style.use('fast')
plt.subplots_adjust(wspace=0.05, hspace=0.05)

for i, c in enumerate(poblacion_region_df.columns):
    poblacion_region_df[c].plot.pie(ax=axs[i])
    axs[i].set_xlabel(c)
    axs[i].set_ylabel('')
    
```

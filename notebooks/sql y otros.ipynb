# SQL y otras características

````python
import pandas as pd
import numpy as np
from random import randint, choice
````

````python
lista_dias = ['Lunes', 'Martes', 'Miércoles','Jueves', 'Viernes','Sábado', 'Domingo']
lista_comidas = ['Desayuno','Almuerzo','Cena','Merienda']
lista_generos = ['F','M']
lista_fumadores = ['N', 'N', 'N', 'S']

facturas = []
totales = []
propinas = []
generos = []
fumadores = []
dias = []
comidas = []
platos = []
````

````python
for i in range(200):
    facturas.append(i)
    totales.append(randint(20, 100))
    propinas.append(randint(1, 10) * 0.75)
    generos.append(choice(lista_generos))
    fumadores.append(choice(lista_fumadores))
    dias.append(choice(lista_dias))
    comidas.append(choice(lista_comidas))
    platos.append(randint(1,4))
````

````python
data = {
    'facturas': facturas,
    'totales': totales,
    'propinas': propinas,
    'generos': generos,
    'fumadores': fumadores,
    'dias': dias,
    'comidas': comidas,
    'platos': platos 
}
````

````python
facturas = pd.DataFrame(data, columns=['facturas','totales','propinas','generos',
                                      'fumadores','dias','comidas','platos'])

facturas.head(10)
````

````python
facturas.columns = ['id','total','propina','genero','fumador','día','momento','cant platos']

facturas.head()
````

# JOINS


````python
df1 = pd.DataFrame(
    {
        "key": ['A', 'B', 'C', 'D'],
        'value': np.random.randn(4)
    }
)

df2 = pd.DataFrame(
    {
        "key": ['B', 'D', 'E', 'D'],
        'value': np.random.randn(4)
    }
)

display(df1, df2)
````

```sql
    SELECT *
    FROM df1
    INNER JOIN df2
    ON df1.key = df2.key    
```

````python
pd.merge(
    df1,
    df2,
    on="key"
)
````

```sql
    SELECT *
    FROM df1
    LEFT OUTER JOIN df2
    ON df1.key = df2.key    
```

````python
pd.merge(
    df1,
    df2,
    on="key",
    how="left"
)
````

```sql
    SELECT *
    FROM df1
    RIGHT OUTER JOIN df2
    ON df1.key = df2.key    
```

````python
pd.merge(
    df1,
    df2,
    on="key",
    how="right"
)
````

```sql
    SELECT *
    FROM df1
    FULL OUTER JOIN df2
    ON df1.key = df2.key    
```

````python
pd.merge(
    df1,
    df2,
    on="key",
    how="outer"
)
````

# UNION

````python
df1 = pd.DataFrame(
    {
        'ciudad': ['Bogota','Cali','Medellin','Chía'],
        'rank': range(1,5)
    }
)

df2 = pd.DataFrame(
    {
        'ciudad': ['Pasto','Tolima','Medellin'],
        'rank': range(1,4)
    }
)

display(df1, df2)
````

```sql
    SELECT ciudad, rank
    FROM df1
    UNION ALL
    SELECT ciudad, rank
    FROM df2
```

````python
pd.concat(
    [df1, df2]
)
````

```sql
    SELECT ciudad, rank
    FROM df1
    UNION DISTINCT
    SELECT ciudad, rank
    FROM df2
```

````python
pd.concat(
    [df1, df2]
).drop_duplicates()
````
